# Convert datetime columns to datetime data type and calculate fire duration in days

```python
# dependancies
import pandas as pd
```

```python
# File to Load
file_path = "static/data/wildfires.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_path)

# display first five rows of dataframe
df.head(50)
```

```python
# check data types
df.dtypes
```

```python
# remove T and add a space
df["start_datetime"] = df["start_datetime"].str.replace(r'T', ' ')
df["start_datetime"]
```

```python
# remove Z
df["start_datetime"] = df["start_datetime"].str.replace(r'Z', '')
df["start_datetime"]
```

```python
# conert column to datatime type
df["start_datetime"] = pd.to_datetime(df["start_datetime"])
df["start_datetime"]
```

```python
# remove T and add a space
df["extinguished_datetime"] = df["extinguished_datetime"].str.replace(r'T', ' ')
df["extinguished_datetime"]
```

```python
# remove Z
df["extinguished_datetime"] = df["extinguished_datetime"].str.replace(r'Z', '')
df["extinguished_datetime"]
```

```python
# conert column to datatime type
df["extinguished_datetime"] = pd.to_datetime(df["extinguished_datetime"])
df["extinguished_datetime"]
```

```python
# verify characters were removed and columns converted to datetime
df.head(5)
```

```python
# calculate fire duration in days 
df['fire_duration'] = df['extinguished_datetime'] - df['start_datetime']
df["fire_duration"]
```

```python
# create new column with number of days from fire duration
df["duration_days"] = df["fire_duration"].dt.days
df["duration_days"]
```

```python
# fillna values with 0 to convert column to integer
df['duration_days'] = df['duration_days'].fillna(0)
df["duration_days"].astype(int)
```

```python
# set decimal precision and remove .0 from days
pd.set_option('precision', 0)
df["duration_days"].round(0)
df["duration_days"].astype(int)
```
